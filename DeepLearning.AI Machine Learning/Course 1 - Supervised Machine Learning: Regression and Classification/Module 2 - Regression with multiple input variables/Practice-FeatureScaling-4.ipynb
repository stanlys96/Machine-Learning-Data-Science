```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_multi import  load_house_data, run_gradient_descent 
from lab_utils_multi import  norm_plot, plt_equal_scale, plot_cost_i_w
from lab_utils_common import dlc
import seaborn as sns
np.set_printoptions(precision=2)
plt.style.use('./deeplearning.mplstyle')
# load the dataset
X_train, y_train = load_house_data()
X_features = ['size(sqft)','bedrooms','floors','age']
# w_norm, b_norm, hist = run_gradient_descent(X_norm, y_train, 1000, 1.0e-1, )
```

```python
n = X_train.shape[1]
fig, ax = plt.subplots(1, n, figsize=(12, 3), sharey=True)
for i in range(n):
    ax[i].scatter(X_train[:, i], y_train)
    ax[i].set_xlabel(X_features[i])
plt.show()
```

```python
def z_normalize_features(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma
```

```python
X_norm, mu, sigma = z_normalize_features(X_train)
```

```python
mu = np.mean(X_train, axis=0)
sigma = np.std(X_train, axis=0)
X_mean = X_train - mu
X_norm = (X_train - mu) / sigma
fig, ax = plt.subplots(1, 3, figsize=(12, 3))

ax[0].scatter(X_train[:, 0], X_train[:, 3])
ax[0].axis('equal')

ax[1].scatter(X_mean[:, 0], X_mean[:, 3])
ax[1].axis('equal')

ax[2].scatter(X_norm[:, 0], X_norm[:, 3])
ax[2].axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i in range(X_train.shape[1]):
    sns.histplot(X_train[:, i], kde=True, ax=ax[i])
plt.show()
fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i in range(X_norm.shape[1]):
    sns.histplot(X_norm[:, i], kde=True, ax=ax[i])
plt.show()
```

```python
w_norm, b_norm, hist = run_gradient_descent(X_norm, y_train, 1000, 1.0e-1, )
py = np.zeros_like(y_train)
for i in range(X_norm.shape[0]):
    py[i] = np.dot(X_norm[i], w_norm) + b_norm
```

```python
fig, ax = plt.subplots(1, X_train.shape[1], figsize=(12, 3), sharey=True)
for i in range(X_train.shape[1]):
    ax[i].scatter(X_train[:, i], y_train, label="actual")
    ax[i].scatter(X_train[:, i], py, c="orange", label="prediction")
ax[0].legend()
plt.show()
```
