```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_multi import zscore_normalize_features, run_gradient_descent_feng
np.set_printoptions(precision=2)  # reduced display precision on numpy arrays
```

```python
x = np.arange(0, 20, 1)
y = 1 + x ** 2
X = x.reshape(-1, 1)
w_out, b_out = run_gradient_descent_feng(X, y, 10000, 1.0e-5)
plt.scatter(x, y, marker="x", c="red", label="Actual")
plt.plot(x, X @ w_out + b_out, label="Prediction")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = 1 + x ** 2
X = x ** 2
X = X.reshape(-1, 1)
w_out, b_out = run_gradient_descent_feng(X, y, 10000, 1.0e-5)
plt.scatter(x, y, marker="x", c="red", label="Actual")
plt.plot(x, np.dot(X, w_out) + b_out, label="Prediction")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x ** 2
X = np.c_[x, x ** 2, x ** 3]
n = X.shape[1]
x_labels = ["x", "x^2", "x^3"]
fig, ax = plt.subplots(1, n, sharey=True, figsize=(12,3))
for i in range(n):
    ax[i].scatter(X[:, i], y)
    ax[i].set_xlabel(x_labels[i])
plt.show()
```
