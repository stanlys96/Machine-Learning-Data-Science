```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from lab_utils_multi import  load_house_data
from lab_utils_common import dlc
np.set_printoptions(precision=2)
plt.style.use('./deeplearning.mplstyle')
```

```python
X_train, y_train = load_house_data()
X_features = ['size(sqft)','bedrooms','floors','age']
```

```python
scaler = StandardScaler()
sgd = SGDRegressor(max_iter=10000)
X_scaled = scaler.fit_transform(X_train)
sgd.fit(X_scaled, y_train)
print(np.ptp(X_train, axis=0))
print(np.ptp(X_scaled, axis=0))
print(sgd.coef_)
print(sgd.intercept_)
```

```python
result = sgd.predict(X_scaled)
```

```python
n = X_train.shape[1]
fig, ax = plt.subplots(1, n, sharey=True, figsize=(12, 3))
for i in range(n):
    ax[i].scatter(X_train[:, i], y_train, label="Actual")
    ax[i].scatter(X_train[:, i], result, label="Prediction", c="orange")
plt.show()
```
