```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])                                           #(m,)
pos = y_train == 1
neg = y_train == 0
```

```python
plt.close("all")
x0 = np.arange(0, 6)
x1 = 3 - x0
plt.scatter(X_train[pos, 0], X_train[pos, 1], marker="x", s=80, c="red")
plt.scatter(X_train[neg, 0], X_train[neg, 1], c="blue", s=100)
plt.plot(x1, x0)
plt.axis([0, 4, 0, 3.5])
plt.show()
```

```python
def current_sigmoid(z):
  g = 1.0 / (1.0 + np.exp(-z))
  return g
```

```python
def compute_logistic_loss(X, y, w, b):
  m = X.shape[0]
  total_cost = 0
  for i in range(m):
    y_pred = np.dot(X[i], w) + b
    total_cost += -y[i] * np.log(current_sigmoid(y_pred)) - (1 - y[i]) * np.log(1 - current_sigmoid(y_pred))
  total_cost /= m
  return total_cost
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
print(compute_logistic_loss(X_train, y_train, w_tmp, b_tmp))
```

```python
w_array1 = np.array([1,1])
b_1 = -3
w_array2 = np.array([1,1])
b_2 = -4

print("Cost for b = -3 : ", compute_logistic_loss(X_train, y_train, w_array1, b_1))
print("Cost for b = -4 : ", compute_logistic_loss(X_train, y_train, w_array2, b_2))
```
