```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from lab_utils_common import plot_data, sigmoid, draw_vthresh
plt.style.use('./deeplearning.mplstyle')
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
plt.close('all')
pos = y == 1
neg = y == 0
plt.scatter(X[pos, 0], X[pos, 1], s=80, marker="x", c="red", label="y=1")
plt.scatter(X[neg, 0], X[neg, 1], s=100, c="blue", label="y=0")
plt.legend()
plt.axis([0, 4, 0, 3.5])
plt.show()
```

```python
def current_sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
z_tmp = np.arange(-10, 11)
exponented = np.c_[z_tmp, current_sigmoid(z_tmp)]
print(exponented)
```

```python
plt.close('all')
plt.plot(exponented[:, 0], exponented[:, 1])
plt.show()
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
pos = y == 1
neg = y == 0
```

```python
plt.close('all')
x0 = np.arange(0, 4)
x1 = 3 - x0
plt.scatter(X[pos, 0], X[pos, 1], s=80, marker="x", c="red", label="y=1")
plt.scatter(X[neg, 0], X[neg, 1], s=100, c="blue", label="y=0")
plt.legend()
plt.plot(x0, x1)
plt.axis([0, 4, 0, 3.5])
plt.show()
```
