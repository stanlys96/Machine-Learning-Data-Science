```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
plt.close('all')
z_tmp = np.arange(-10, 11)
exponented = np.c_[z_tmp, sigmoid(z_tmp)]
plt.plot(exponented[:, 0], exponented[:, 1])
plt.xticks(z_tmp)
plt.grid()
plt.show()
```
